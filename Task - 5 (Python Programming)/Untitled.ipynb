```python
#Question-1 (Detect Floating Point Number)

import re
n=int(input())
pattern=r'^[+-]?[0-9]*\.[0-9]+$'
for i in range(n):
    s=input()
    print(bool(re.match(pattern,s)))
```

```python
#Question-3 (Group(), Groups() & Groupdict())

import re
s=input()
pattern=r'([a-z A-Z 0-9])\1'
m=re.search(pattern,s)
if m:
    print(m.groups()[0])
else:
    print(-1)
```

```python
#Question-4 (Re.findall() & Re.finditer())

import re
s=input()
v="aeiou"
c="qwrtypsdfghjklzxcvbnm"
l=re.findall(r"(?<=[%s])([%s]{2,})[%s]" % (c,v,c),s,flags=re.IGNORECASE)
if not l:
    print(-1)
else:
    for i in l:
        print(i)
```

```python
#Question-5 (Re.start() & Re.end())

import re
s=input()
k=input()
pattern=re.compile(k)
m=pattern.search(s)
if not m:
    print("(-1, -1)")
else:
    while m:
        print("({0}, {1})".format(m.start(), m.end() - 1))
        m=pattern.search(s, m.start() + 1)
```

```python
#Question-6 (Regex Substitution)

import re

pattern = r"(?<= )(&&|\|\|)(?= )"
replacement = lambda x: "and" if x.group() == "&&" else "or"

for i in range(int(input())):
    s = input()
    print(re.sub(pattern, replacement, s))
```

```python
#Question-7 (Validating Roman Numerals)

thousands = 'M{0,3}'
hundreds = '(CM|CD|D?C{0,3})'
tens = '(XC|XL|L?X{0,3})'
digits = '(IX|IV|V?I{0,3})'

regex_pattern = r"%s%s%s%s$" % (thousands, hundreds, tens, digits)	# Do not delete 'r'.

import re
print(str(bool(re.match(regex_pattern, input()))))
```

```python
#Question-8 (Validating phone numbers)

import re
def checker(contact):
    pattern = r"[789]\d{9}$"
    if re.match(pattern, contact):
        return "YES"
    else:
        return "NO"
    
n = int(input())
for i in range(n):
    print(checker(input()))
```

```python
#Question-9 (Validating and Parsing Email Addresses)

import re
n=int(input())
for i in range(n):
    line = input()
    name,email = line.split(" ")
    pattern = "<[a-z][a-zA-Z0-9\-\._]+@[a-zA-Z]+\.[a-zA-Z]{1,3}>"
    if bool(re.match(pattern,email)):
        print(name,email)
```

```python
#Question-10 (Hex Color Code)

import re
in_css = False
n=int(input())
for _ in range(n):
    line = input()
    if '{' in line:
        in_css = True
    elif '}' in line:
        in_css = False
    elif in_css:
        for color in re.findall('#[0-9a-fA-F]{3,6}',line):
            print(color)
```

```python
#Question-11 (HTML Parser - Part 1)

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print("Start :", tag)
        self.print_attrs(attrs)
    def handle_endtag(self, tag):
        print("End   :", tag)
    def handle_startendtag(self, tag, attrs):
        print("Empty :", tag)
        self.print_attrs(attrs)
        
    def print_attrs(self, attrs):
        for elem in attrs:
            print("-> {:} > {:}".format(elem[0], elem[1]))
            
if __name__ == "__main__":
    parser = MyHTMLParser()
    for _ in range(int(input())):
        parser.feed(input())
```

```python
#Question-12 (HTML Parser - Part 2)

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        if data != '\n':
            if '\n' not in data:
                print(">>> Single-line Comment")
                print(data)
            else:
                print(">>> Multi-line Comment")
                print(data)
    
    def handle_data(self, data):
        if data != '\n':
            print(">>> Data")
            print(data)
  
  
  
  
   
html = ""       
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#Question-13 (Detect HTML Tags, Attributes and Attribute Values)

from html.parser import HTMLParser
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print(tag)
        for at in attrs:
            print("-> {} > {}".format(at[0], at[1]))
            
    def handle_startendtag(self, tag, attrs):
        print(tag)
        for at in attrs:
            print("-> {} > {}".format(at[0], at[1]))
            
html = ""
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#Question-14 (Validating UID)

import re

for i in range(int(input())):
    N = input().strip()
    if N.isalnum() and len(N) == 10:
        if bool(re.search(r'(.*[A-Z]){2,}',N)) and bool(re.search(r'(.*[0-9]){3,}',N)):
            if re.search(r'.*(.).*\1+.*',N):
                print('Invalid')
            else:
                print('Valid')
        else:
            print('Invalid')
    else:
        print('Invalid')
```

```python
#Question-15 (Validating Credit Card Numbers)

import re

for _ in range(int(input())):
    n = input().strip()
    try:
        assert re.match(r'(\d{4}-){3}\d{4}$', n) or re.match(r'\d{16}$', n)
        n = re.sub(r'-', '', n)
        assert re.match(r'[456]', n)
        assert not re.search(r'(\d)\1{3,}', n)
    except:
        print('Invalid')
    else:
        print('Valid')
```
